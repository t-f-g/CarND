```python
# Preprocess Images with opencv2
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline
 
# load the image and show it
#reading in an image
image = mpimg.imread("data\IMG\center_2016_12_01_13_30_48_287.jpg")
plt.imshow(image)
```

```python
image_crop = image[40:40+96]
#print(image_crop.shape[0])
plt.imshow(image_crop)
```

```python
def convert_to_hsv(img):
        """
        return HSV version of image
        """
        return cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

image_crop_hsv = convert_to_hsv(image_crop)
plt.imshow(image_crop_hsv)
```

```python
def hsv_value(img):
        """
        return value channel of HSV color space
        """
        return img[:, :, 2]

image_crop_hsv_v = hsv_value(image_crop_hsv)
plt.imshow(image_crop_hsv_v, cmap='gray')
```

```python
#optional
def equalize(img):
        """
        return equalized image
        """
        return cv2.equalizeHist(img)
    
image_crop_hsv_v_eq = equalize(image_crop_hsv_v)
plt.imshow(image_crop_hsv_v_eq, cmap='gray')
```

```python
# resize
image_crop_hsv_v_eq_square = cv2.resize(image_crop_hsv_v_eq, (32,32), interpolation = cv2.INTER_AREA)
plt.imshow(image_crop_hsv_v_eq_square, cmap='gray')
```
